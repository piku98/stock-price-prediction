```python
data_df = df.drop(['LABEL', 'Date'], axis=1)
```

```python
n = data_df.shape[0]
```

```python
X_df = data_df.drop(['Close'], axis=1)
y_df = data_df['Close']
X_data = X_df.values
y_data = y_df.values
```

```python
X_train = X_data[0:int(n*0.8), :]
y_train = y_data[0:int(n*0.8)]
X_test = X_data[int(n*0.8):, :]
y_test = y_data[int(n*0.8):]
```

```python
n_input = 11
n_l1 = 1024
n_l2 = 512
n_l3 = 256
n_out = 1
```

```python
X = tf.placeholder(dtype=tf.float32, shape=[None, n_input])
Y = tf.placeholder(dtype=tf.float32, shape=[None])
```

```python
sigma = 1
weight_initializer = tf.variance_scaling_initializer(mode="fan_avg", distribution="uniform", scale=sigma)
bias_initializer = tf.zeros_initializer()
```

```python
w1 = tf.Variable(weight_initializer([n_input, n_l1]))
b1 = tf.Variable(bias_initializer([n_l1]))
w2 = tf.Variable(weight_initializer([n_l1, n_l2]))
b2 = tf.Variable(bias_initializer([n_l2]))
w3 = tf.Variable(weight_initializer([n_l2, n_l3]))
b3 = tf.Variable(bias_initializer([n_l3]))
w_out = tf.Variable(weight_initializer([n_l3, n_out]))
b_out = tf.Variable(bias_initializer([n_out]))
```

```python
l1 = tf.nn.relu(tf.add(tf.matmul(X, w1), b1))
l2 = tf.nn.relu(tf.add(tf.matmul(l1, w2), b2))
l3 = tf.nn.relu(tf.add(tf.matmul(l2, w3), b3))
out = tf.transpose(tf.add(tf.matmul(l3, w_out), b_out))
Y.shape
```

```python
mse = tf.reduce_mean(tf.squared_difference(out, Y))
```

```python
opt = tf.train.AdamOptimizer().minimize(mse)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(100):
        sess.run(opt, feed_dict={X: X_train, Y: y_train})
        if i % 10 == 0:
            test_out = sess.run(out, feed_dict={X: X_train})
            #mse = sess.run(mse, feed_dict={X: X_train, Y: y_train})
            print(test_out)
```
