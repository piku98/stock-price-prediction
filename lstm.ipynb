```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,LSTM,Dropout
```

```python
df = pd.read_csv('NASDAQ.csv')
```

```python
df.head()
```

```python
data_df = df.drop(['Date', 'LABEL'], axis=1)
```

```python
data_df.head()
```

```python
data_df_X = data_df.drop(['Close'], axis=1)[:-1]
data_df_y = data_df['Close']
data_df_y = data_df_y.shift(-1, axis=0)[:-1]
X = data_df_X.values
y = data_df_y.values
```

```python
corr = data_df.corr()
```

```python
corr
```

```python
sb.set(rc={'figure.figsize':(11.7,8.27)})
sb.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=False)
y_train = y_train.reshape(-1, 1)
y_test = y_test.reshape(-1, 1)
```

```python
scaler = MinMaxScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.fit_transform(X_test)
y_train_norm = scaler.fit_transform(y_train)
y_test_norm = scaler.fit_transform(y_test)
```

```python
X_train_norm = np.reshape(X_train_norm, (X_train_norm.shape[0], 1, X_train_norm.shape[1]))
X_test_norm = np.reshape(X_test_norm, (X_test_norm.shape[0], 1, X_test_norm.shape[1]))
```

```python
X_train_norm.shape, X_test_norm.shape
```

```python
regressior = Sequential()

regressior.add(LSTM(units = 50, activation = 'relu', return_sequences = True, input_shape =(1,11)))
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
regressior.add(Dropout(0.2))

regressior.add(LSTM(units = 120, activation = 'relu', return_sequences = True))
regressior.add(Dropout(0.2))

regressior.add(Dense(units = 1))
```

```python
regressior.summary()
```

```python
regressior.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressior.fit(X_train_norm, y_train_norm, epochs = 10, batch_size = 19)
```

```python
y_pred = regressior.predict(X_test_norm)
```
