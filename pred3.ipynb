```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.figure as figure
```

```python
df = pd.read_csv('NASDAQ.csv')
```

```python
df.head()
```

```python
data_df = df.drop(['Date', 'LABEL'], axis=1)
data_df.head()
```

```python
data_df_X = data_df.drop(['Close'], axis=1)[: -1]
data_df_y = data_df['Close'].shift(-1, axis=0)[ : -1]
X = data_df_X.values
y = data_df_y.values
y = y.reshape(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=False)
X_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
predictions = lr.predict(X_test)
```

```python
plt.plot(range(len(y_test)), y_test.flatten(), color='b')
plt.plot(range(len(y_test)), predictions, color='r')
figsize=(11.7, 8.27)
plt.show()
```
